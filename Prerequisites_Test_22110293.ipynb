Question 2 

```python
#using list of lists

import time

def matrix_mult_list(A, B):
    n = len(A)
    k = len(A[0])
    m = len(B[0])
    C = [[0] * m for _ in range(n)]
    
    for i in range(n):
        for j in range(m):
            for r in range(k):
                C[i][j] += A[i][r] * B[r][j]
                
    return C

A = [[1, 2], [3, 4], [5, 6]]
B = [[7, 8, 9], [10, 11, 12]]

start_time = time.time()
C_list = matrix_mult_list(A, B)
end_time = time.time()
list_duration = end_time - start_time

print("Result using list of lists:\n", C_list)
print("Time taken using list of lists:", list_duration)
```

```python
## using numpy

import numpy as np
A = [[1, 2], [3, 4], [5, 6]]
B = [[7, 8, 9], [10, 11, 12]]
start_time = time.time()
A_np = np.array(A)
B_np = np.array(B)
C_np = np.dot(A_np, B_np)
end_time = time.time()
numpy_duration = end_time - start_time

print("Result using numpy:", C_np)
print("Time taken using numpy:", numpy_duration)
```

Matrix multiplication is quicker with numpy compared to using list of lists.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0.5, 100.0, 0.5)

y1 = x
y2 = x**2
y3 = x**3 / 100
y4 = np.sin(x)
y5 = np.sin(x) / x
y6 = np.log(x)
y7 = np.exp(x)
plt.figure(figsize=(14, 10))

#  y = x
plt.subplot(3, 3, 1)
plt.plot(x, y1, label='y = x')
plt.title('y = x')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

#  y = x^2
plt.subplot(3, 3, 2)
plt.plot(x, y2, label='y = x^2', color='orange')
plt.title('y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

# y = x^3 / 100
plt.subplot(3, 3, 3)
plt.plot(x, y3, label='y = x^3 / 100', color='green')
plt.title('y = x^3 / 100')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

# y = sin(x)
plt.subplot(3, 3, 4)
plt.plot(x, y4, label='y = sin(x)', color='red')
plt.title('y = sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

# y = sin(x) / x
plt.subplot(3, 3, 5)
plt.plot(x, y5, label='y = sin(x) / x', color='purple')
plt.title('y = sin(x) / x')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

# y = log(x)
plt.subplot(3, 3, 6)
plt.plot(x, y6, label='y = log(x)', color='brown')
plt.title('y = log(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

# y = e^x
plt.subplot(3, 3, 7)
plt.plot(x, y7, label='y = e^x', color='pink')
plt.title('y = e^x')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim([0, 1000])  
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

Question 10

```python
import numpy as np
import pandas as pd

matrix = np.random.uniform(low=1.0, high=2.0, size=(20, 5))

columns = ['a', 'b', 'c', 'd', 'e']
df = pd.DataFrame(matrix, columns=columns)
std_devs = df.std()
max_std_col = std_devs.idxmax()
max_std_value = std_devs.max()
row_means = df.mean(axis=1)
min_mean_row_index = row_means.idxmin()
min_mean_value = row_means.min()
print("DataFrame:")
print(df)
print("\nColumn with the highest standard deviation:")
print(f"Column: {max_std_col}, Standard Deviation: {max_std_value:.4f}")
print("\nRow with the lowest mean:")
print(f"Row Index: {min_mean_row_index}, Mean: {min_mean_value:.4f}")
```

Question 13

```python
import numpy as np

def compute_argmin(arr):
    arr = np.asarray(arr)
    min_index = 0
    
    for i in range(len(arr)):
        if arr[i] < arr[min_index]:
            min_index = i
    
    return min_index
    
array = np.array([5, 2, 9, 1, 7])
custom_argmin = compute_argmin(array)
numpy_argmin = np.argmin(array)

print(f"Custom argmin: {custom_argmin}")
print(f"NumPy argmin: {numpy_argmin}")
```
